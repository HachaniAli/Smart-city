```python
# Importation de librairies

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import DBSCAN
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.ExcelFile("Unicorn.xlsx") # Load spreadsheet
df = df.parse('Unicorns', skiprows=2)
```

```python
df = df[['Company', 'Valuation ($B) ', 'Date Added', 'Country',
       'Category', 'Select Investors']]
df.rename(columns={"Valuation ($B) ": "Valuation", "Date Added": "Date_Added", "Select Investors": "SelectInvestors"}, inplace=True)
```

```python
India = df[df['Country'].str.contains('India')]
India = India.sort_values(by='Date_Added', ascending=False)
India.reset_index(inplace=True)
India = India[['Company','Valuation', 'Date_Added','Country','Category','SelectInvestors']]
India.head(3)
```

# Investors

```python
India['SelectInvestors'] = India['SelectInvestors'].astype('str')
```

```python
investors = []

for i in range(len(India)):
    for j in range(len(India.loc[i, 'SelectInvestors'].split(','))):
        investors.append(India.loc[i, 'SelectInvestors'].split(',')[j])

output = []
for x in investors:
    if x not in output:
        output.append(x)

investors = output

# 2nd piece of code


my_dict = {}
for i in range(len(investors)):
    my_dict[investors[i]] = 0
    
    
#3 rd piece of code

for i in range(len(India)):
    for j in range(len(India.loc[i, 'SelectInvestors'].split(','))):
        for key in my_dict:
            if India.loc[i, 'SelectInvestors'].split(',')[j] == key:
                my_dict[key] += 1
                
#4th piece of code

data_items = my_dict.items()
data_list = list(data_items)
#data_list
df_india = pd. DataFrame(data_list) 
df_india
```
